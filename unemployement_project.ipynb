```python
fig = px.line(df_2020,x= ' Date',y= ' Estimated Unemployment Rate (%)',title='Unemployment rate over time ',labels={' Estimated Unemployment Rate (%)':'Unemploy rate'})#,template='plotly_dark' 

covid = {'start_date':'2020-03-01','end_date': '2021-06-30','event':'peak of pendemic'}
fig.add_shape(type='rect', x0=pd.to_datetime(covid['start_date']), 
              x1=pd.to_datetime(covid['end_date']),y0=0,y1=1, 
              fillcolor='red', opacity=0.4, layer='below') 

#fig.update_layout if we want to update in xaxis and yaxis name and change title 
fig.update_xaxes(type='category')
fig.show()
            
```

```python
fig = px.bar(df_2020, x= 'Region',y=' Estimated Unemployment Rate (%)',color='Region',
             animation_frame=' Date' , title='unemployement rate by time ',
             labels={' Estimated Unemployment Rate (%)' : 'Unemployement rate' },
             template='plotly', color_discrete_sequence= px.colors.sequential.Rainbow,
             category_orders={'Region': ['Region1','Region2','Region3',]},height=500)

fig.update_layout(xaxis= {'categoryorder' : 'total descending'},barmode= 'stack',
                  showlegend= False)

'''fig.add_annotation (text='impact over time',
                    xref='paper',yref='paper',
                    x=0.5, y=1.05,
                    showarrow=False,
                    font=dict(size=16, color= 'black')) '''
fig.show()
```
